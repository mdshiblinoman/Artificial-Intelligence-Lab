```python
import torch
from torchvision import models, transforms
from PIL import Image
```

```python
# Load pretrained VGG16
weights = models.VGG16_Weights.DEFAULT
model = models.vgg16(weights=weights)

# Set the model to evaluation mode
model.eval()
```

```python
preprocess = weights.transforms()
```

```python
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),
])
```

```python
img = Image.open("cat.jpg").convert("RGB")

input_tensor = preprocess(img)
input_batch = input_tensor.unsqueeze(0)  # Shape: (1, 3, 224, 224)
```

```python
with torch.no_grad():
    output = model(input_batch)
```

```python
probabilities = torch.nn.functional.softmax(output[0], dim=0)

predicted_class = probabilities.argmax().item()

print("Predicted Class Index:", predicted_class)
print("Confidence:", probabilities[predicted_class].item())
```

```python
categories = weights.meta["categories"]

print("Prediction:", categories[predicted_class])
print("Confidence: {:.2f}%".format(probabilities[predicted_class].item() * 100))
```
